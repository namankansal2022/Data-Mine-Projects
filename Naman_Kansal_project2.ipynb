```r
sum(myDF$genre == "Comedy")
sum(grepl("Comedy", myDF$genre) & myDF$genre != "Comedy")
sum(grepl("Comedy", myDF$genre))
sort(table(myDF$genre[grepl("Comedy", myDF$genre) & myDF$genre != "Comedy"]), decreasing = TRUE)[1]
sum(myDF$genre == "Romance")
sum(grepl("Romance", myDF$genre) & myDF$genre != "Romance")
sum(grepl("Romance", myDF$genre))
```

**Question 3: Genre Analysis**

**sum(myDF$genre == "Comedy")** counts movies exclusively labeled **"Comedy"**.

- **sum(grepl("Comedy", myDF$genre) , myDF$genre != "Comedy")** counts movies with multiple genres including **"Comedy"**.

- **sum(grepl("Comedy", myDF$genre))** counts all movies containing **"Comedy"**.

- **sort(table(myDF$genre[grepl("Comedy", myDF$genre) & myDF$genre != "Comedy"]), decreasing = TRUE)[1]** identifies the most common multi-genre **"Comedy"** category.

- The same steps are applied for **"Romance"** using **==** and **grepl()** to get exclusive, multi-genre, and total counts.
