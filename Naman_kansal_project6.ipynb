```r
screen_time <- 10

while(screen_time > 2) {
    print(paste("Screen time:", screen_time, "hours"))
    screen_time <- screen_time - 1
}

cat("\n")  

sleep_time <- 2

while(sleep_time < 10) {
    print(paste("Sleep time:", sleep_time, "hours"))
    sleep_time <- sleep_time + 1
}

cat("\n") 

screen_time <- 5
sleep_time <- 6

while (screen_time > 2) {
    print(paste("Screen time:", screen_time, "hours"))
    print(paste("Sleep time:", sleep_time, "hours"))
    
    screen_time <- screen_time - 0.5
    sleep_time <- sleep_time + 0.5
}

```

# Explanation of Code

- **Line 1**: Initializes `screen_time` with a starting value of 10.  
- **Lines 3–6**: Runs a `while` loop that continues as long as `screen_time` is greater than 2. Each iteration prints the current `screen_time` value and decreases it by 1 hour.  
- **Line 8**: Uses `cat("\n")` to add a blank line between loop outputs for better readability.  

- **Line 10**: Initializes `sleep_time` with a starting value of 2.  
- **Lines 12–15**: Executes a second `while` loop that continues as long as `sleep_time` is less than 10. Each iteration prints the current `sleep_time` value and increases it by 1 hour.  
- **Line 17**: Inserts another line break between outputs using `cat("\n")`.  

- **Line 19–20**: Resets `screen_time` to 5 and `sleep_time` to 6 before the final loop begins.  
- **Lines 22–28**: Runs a final `while` loop that continues while `screen_time` is greater than 2. It prints both variables in each iteration, then decreases `screen_time` by 0.5 and increases `sleep_time` by 0.5 each time.  
- This loop demonstrates how both values change together until the stopping condition is reached.

  
- `paste()` creates a character vector, and `print()` displays that vector with R’s default formatting.  
- `cat()` directly concatenates and prints its arguments as plain text, without quotes, indexes, or line breaks unless explicitly added.
