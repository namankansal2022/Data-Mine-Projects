```r
top_20 <- names(head(sort(table(myDF$Dest), decreasing = TRUE),20))
top_20
top_dests_df <- myDF[myDF$Dest %in% top_20, ]
tapply(top_dests_df$AirTime, list(top_dests_df$Origin, top_dests_df$Dest), mean, na.rm=TRUE)
```

# Explanation of Code

- **Line 1**: Creates `top_20` by making a frequency table of destinations (`table(myDF$Dest)`), sorting it in descending order with `sort()`, keeping the top 20 with `head()`, and extracting the airport codes using `names()`.  
- **Line 2**: Displays the contents of `top_20`, which are the codes of the 20 busiest destination airports.  
- **Line 3**: Creates a subset dataframe `top_dests_df` that includes only flights whose destination airport is in the `top_20` list.  
- **Line 4**: Uses `tapply()` with two grouping factors (`Origin`, `Dest`) to compute the average `AirTime` for each origindestination pair in `top_dests_df`, removing missing airtime values with `na.rm = TRUE`.  

# Deliverable Answers

- **2.1**: The airports in `top_20` are the 20 most frequent destinations in the dataset (the specific list of airport codes is displayed when `top_20` is printed).  
- **2.2**: We used `sort()` before `names()` so that we extracted the top 20 *busiest destinations* in order of frequency. Without sorting, `names()` would have returned the airport codes in default (alphabetical) order rather than by flight volume.  
- **2.3**: The `NA` values in the OriginDestination "table" appear because there are **no flights recorded** for those pairs in the dataset. Even though `na.rm = TRUE` removes missing `AirTime` values when calculating averages, it cannot invent values for pairs of airports that simply did not have flights. So `NA` means this route doesnt exist in the 2006 data, not that the airtime values were missing.

## Question 4

```r
my_flights <- read.csv("/anvil/projects/tdm/data/flights/subset/1997.csv")
depart_times <- cut(
  my_flights$DepTime,
  breaks = c(0, 300, 600, 900, 1200, 1500, 1800, 2100, 2400),
  labels = c("Night", 
             "Early Morning", 
             "Morning", 
             "Late Morning",
             "Afternoon",
             "Late Afternoon",
             "Evening",
             "Late Evening"),
  include.lowest = TRUE
)

table(depart_times)

barplot(table(depart_times),
        main = "Flight Departures by Time of Day (1997)",
        xlab = "Time of Day",
        ylab = "Number of Flights",
        col = c("red", "blue", "pink", "lightblue", 
                "orange", "yellow", "green", "violet"),
        las = 2,
        cex.names = 0.8)

arrival_times <- cut(
  my_flights$ArrTime,
  breaks = c(0, 300, 600, 900, 1200, 1500, 1800, 2100, 2400),
  labels = c("Night", 
             "Early Morning", 
             "Morning", 
             "Late Morning",
             "Afternoon",
             "Late Afternoon",
             "Evening",
             "Late Evening"),
  include.lowest = TRUE
)

table(arrival_times)

barplot(table(arrival_times),
        main = "Flight Arrivals by Time of Day (1997)",
        xlab = "Time of Day",
        ylab = "Number of Flights",
        col = c("red", "blue", "pink", "lightblue", 
                "orange", "yellow", "green", "violet"),
        las = 2,
        cex.names = 0.8)


```

# Explanation of Code

- **Line 1**: Reads the 1997 flights dataset into a dataframe named `my_flights`.  
- **Line 210**: Uses `cut()` on the `DepTime` column to group departure times into 8 time-of-day categories (Night, Early Morning, Morning, etc.). The result is stored in a new variable `depart_times`.  
- **Line 11**: Creates a frequency table of `depart_times`, showing the count of flights in each time-of-day group.  
- **Line 1218**: Creates a barplot of `depart_times` frequencies, with appropriate title, axis labels, colors, and rotated x-axis labels for readability.  
- **Line 1927**: Uses `cut()` again on the `ArrTime` column to group arrival times into the same 8 categories. The result is stored in a new variable `arrival_times`.  
- **Line 28**: Creates a frequency table of `arrival_times`, showing the count of flights in each time-of-day group.  
- **Line 2935**: Creates a barplot of `arrival_times` frequencies, again with title, axis labels, colors, and rotated labels.  

# Deliverable Answers

- **4.1**: Using the departure times barplot, the **Morning (06:0109:00)** time of day had the most departure flights (over 1 million).  
- **4.2**: Using the arrival times barplot, the **Early Morning (03:0106:00)** time of day had the least arrival flights (only ~38,000).  
- **4.3**: From this analysis, we can conclude that most flights are scheduled to **depart in the morning** and **arrive later in the day (Afternoon and Evening)**. Very few flights depart or arrive overnight, which makes sense given typical airline scheduling and passenger demand patterns.
