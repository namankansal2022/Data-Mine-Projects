### 2.1 How are typeof(), class(), and mode() different, and which do you prefer?
- `typeof()` shows the **internal storage type** of an object (e.g., "integer", "double", "character").  
- `class()` shows the **high-level classification** used by R for object behavior (e.g., "data.frame", "factor").  
- `mode()` is more general and older, describing how the object is intended to be used (e.g., "numeric", "character").  
- I prefer using `class()` in most cases, since it aligns better with how R functions treat the object.  

### 2.2 How else we couldve dealt with the NA values

- For larger datasets, functions from the `tidyr` or `dplyr` packages (e.g., `drop_na()`) provide cleaner ways to handle missing data.
