```r
# code here
library(lubridate)
flights <- read.csv("/anvil/projects/tdm/data/flights/subset/1997.csv")
flights$full_dates <- ymd(paste(flights$Year, flights$Month, flights$DayofMonth, sep="-"))
flights$DepTime[flights$DepTime == 2400] <- 0
flights$depHour <- floor(flights$DepTime / 100)
flights$depMinute <- (flights$DepTime / 100 - flights$depHour) * 100
flights$depHour <- as.integer(flights$depHour)
flights$depMinute <- as.integer(round(flights$depMinute))
date_times <- make_datetime(flights$Year,flights$Month, flights$DayofMonth,  flights$depHour, flights$depMinute)
head(date_times)
```

# Explanation of Code

- **Line 2:** Reads the flight data from the given CSV file and stores it in a data frame named `flights`.

- **Line 3:** Creates a new column `full_dates` by combining the `Year`, `Month`, and `DayofMonth` columns into a single date format using `ymd()`.

- **Line 4:** Replaces any departure times equal to `2400` with `0`, since `2400` represents midnight and needs to roll over to the next day.

- **Line 5:** Extracts the hour portion of the departure time by dividing `DepTime` by 100 and taking the integer part using `floor()`.

- **Line 9:** Combines `Year`, `Month`, `DayofMonth`, `depHour`, and `depMinute` into full datetime values using `make_datetime()`, and stores them in the variable `date_times`.

**3.3 What does ymd from the ymd() function actually stand for and what does this function do?**  

`ymd` stands for **Year-Month-Day**.  
The `ymd()` function is a method often used with `paste()` to easily combine three columns  a year column, a month column, and a day column  into one `Date` column containing values in the **YYYY-MM-DD** format.

```r
bostonDF <- flights[flights$Origin == "BOS", ]
boston_distance <- tapply(bostonDF$Distance, bostonDF$Month, mean)
plot(boston_distance, type = "b", main = "BOS Flights", xlab = "Month", ylab = "Average Distance",col='blue')

phxDF <- flights[flights$Origin == "PHX", ]
phoenix_distance <- tapply(phxDF$Distance, phxDF$Month, mean)
plot(phoenix_distance, type = "b", main = "PHX Flights", xlab = "Month", ylab = "Average Distance", col='orange')

mdwDF <- flights[flights$Origin == "MDW", ]
chicago_distance<- tapply(mdwDF$Distance, mdwDF$Month, mean)
plot(chicago_distance, type = "b", main = "MDW Flights", xlab = "Month", ylab = "Average Distance",col='red')

seaDF <- flights[flights$Origin == "SEA", ]
seattle_distance <- tapply(seaDF$Distance, seaDF$Month, mean)
plot(seattle_distance, type = "b", main = "SEA Flights", xlab = "Month", ylab = "Average Distance",col='green')

plot(boston_distance,
    type='b',
    col='blue',
     ylim=c(300,
            1200),
     main='Average Distance From Boston, Phoenix, Chicago, and Seattle',
     xlab='Months',
     ylab='Average Distance'
    )
lines(phoenix_distance,
      type='b',
    col='orange'
      )
lines(chicago_distance,
      type='b',
    col='red'
      )
lines(seattle_distance,
      type='b',
    col='green'
      )
```
