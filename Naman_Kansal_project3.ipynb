```r
# code here
unique(app_store$Content.Rating)
app_store$Upper.Content.Rating = lapply(app_store$Content.Rating, toupper)
unique(app_store$Upper.Content.Rating)
sapply(app_store, class)
lapply(app_store, class)
head(sapply(app_store$Content.Rating,nchar),30)
head(lapply(app_store$Content.Rating,nchar),30)
```

* **Line 1** uses the **unique()** function on the `Content.Rating` column to display all the distinct categories of content ratings without duplicates.  

* **Line 2** creates a new column called **Upper.Content.Rating** by applying the **toupper()** function to each value in `Content.Rating` using **lapply()**, converting all the values to uppercase. Since **lapply()** always returns a list, this new column is stored as a list-column.  

* **Line 3** again uses **unique()**, but now on `Upper.Content.Rating`, to confirm the distinct uppercase values after conversion.  

* **Line 4** uses **sapply()** with the **class()** function to get the class (data type) of each column in the dataframe. Because **sapply()** simplifies results, the output is displayed as a clean vector.  

* **Line 5** also uses the **class()** function, but with **lapply()**, so the output is shown as a list. Both give the same information, but the structure of the result is different.  

* **Line 6** applies the **nchar()** function (which counts the number of characters in a string) to every value in `Content.Rating` using **sapply()**. The result is simplified into a numeric vector, and **head(..., 30)** shows the first 30 values.  

* **Line 7** repeats the same operation with **lapply()**, but this time the output is returned as a list. Again, **head(..., 30)** shows the first 30 elements.

* Another function from the apply family is mapply(). It works like sapply(), but it can take multiple vectors or lists as input and apply a function element by element across them. For example, if you want to add elements from two different vectors together, mapply(sum, x, y) would pair them up and compute the result. This makes mapply() very useful when you need to loop over more than one input at the same time.
